## 项目架构
### Poetry
该项目使用Poetry作为虚拟环境和包管理工具，
- 安装poetry
> window环境
1. 命令行安装poetry
```bash
(Invoke-WebRequest -Uri https://install.python-poetry.org -UseBasicParsing).Content | python -   
```
2. 配置路径
在powershell中调用 `$env:PATH += ";C:\Users\ROG\AppData\Roaming\Python\Scripts" `
在git bash中调用 `export PATH="$PATH:/c/Users/ROG/AppData/Roaming/Python/Scripts"`

`source ~/.bashrc`

- 初始化项目
`poetry init`
